```python
import requests
from bs4 import BeautifulSoup

url = "https://web.archive.org/web/20220101112945/http://ippan.org.np/ippan-members"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')
    
    #Find all sections containing company information
    company_sections = soup.find_all('div', class_='card-body')

    #Loop through each company section
    for company_section in company_sections:
        
        company_info = {}
        
        spans = company_section.find_all('span', style='background-color:#99ffff')

        for span in spans:
            # Extract field name from the <strong> tag
            field_name = span.find('strong')
            if field_name:
                field_name = field_name.get_text().strip().rstrip(':')
                # Extract the value, removing the field name part
                value = span.get_text().replace(field_name, '', 1).strip().strip('"')
                company_info[field_name] = value
                        
            
        # Print the extracted information for each company
        for field, value in company_info.items():
            print(f"{field}: {value}")

        print("===")  # Separating information for different companies
    
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
